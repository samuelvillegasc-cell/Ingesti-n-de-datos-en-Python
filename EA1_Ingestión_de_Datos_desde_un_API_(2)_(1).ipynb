```python
import requests # importacion de la libreria Requests para enviar peticion a la API https://jsonplaceholder.typicode.com/users
import sqlite3 # importacion de la interface para conexion y creacion de bases de datos SQLite
import pandas as pd # importacion de Pandas, libreria para el analisis de datos en Python

#¿Qué es una API?
#Una API es un conjunto de reglas y protocolos que permiten que dos aplicaciones o sistemas de software se comuniquen entre sí, facilitando el intercambio de información de forma ordenada y segura.

#¿Cuál es la función de una API en el desarrollo de software?
#La función de una API en el desarrollo de software es permitir agilizar el desarrollo de aplicaciones, ya que facilita la integración de servicios ya existentes. Además, permite la reutilización de código, lo que contribuye a reducir el tiempo y los costos del dearrollo.

#¿Qué son las APIs REST?
#API REST (Representational State Transfer) es una arquitectura o serie de principios y restricciones que permiten que dos sistemas completamente diferentes, se comuniquen entre sí mediante la web, de manera de manera eficiente.

#Características principales de las APIs REST:
#La arquitectra cliente-servidor: es uno de los principios de las APIs REST. En este modelo, el cliente y el servidor funcionan de manera independiente: el cliente envía solicitudes y el servidor se encarga de procesarlas y devolver una respuesta. El servidor no necesita conocer las características internas del cliente, sino únicamente las solicitud que este realiza.
#Sin estado: la solicitud del cliente al servidor debe contener toda la información necesaria para ser procesada. Además, al no existir una "sesión" la información del usuario no queda registrada, teniendo repetirse el proceso, cada vez que se solicita.
#La capacidad de Caché: obliga al servidor a declarar explícitamente si una respuesta puede almacenarse temporalmente o no. Al clasificar los datos de esta forma, el cliente reutiliza información previa sin repetir la consulta, lo que optimiza los recursos y acelera drásticamente la navegación.
#Sistemas en capas: Esta característica indica que el cliente no puede distinguir si está conectado directamente al servidor final o a un intermediario. La solicitud atraviesa diversas capas que trabajan de forma invisible para garantizar la seguridad, gestionar el tráfico de manera eficaz y permitir que el sistema crezca sin que el usuario note cambios.
#Identificación de recursos: En una API REST, cada elemento del sistema (como un usuario o producto) es un recurso y tiene una dirección única llamada URI. El cliente interactúa con ese recurso mediante representaciones, normalmente en JSON o XML. Además, cada mensaje contiene la información necesaria para entenderse (mensaje autodescriptivo) y el servidor puede incluir enlaces que indican qué acciones se pueden hacer después (HATEOAS).
#Código bajo demanda: Significa que el servidor puede enviar código al cliente para que lo ejecute, ampliando temporalmente lo que el cliente puede hacer. Por ejemplo, el servidor puede enviar scripts de JavaScript para que el navegador los ejecute y así agregar nuevas funciones o comportamientos.

#¿Cómo permiten las APIs la comunicación entre diferentes sistemas o aplicaciones?
#Como mencioné ahorita, las APIs actuan como intermediarias que permiten la comunicación entre diferentes sitemas de software mediante reglas y protocolos definidos.
#Primeramente, el usuario o cliente envía una solicitud al servidor a través de un endpoint utilizando métodos como GET (protocolo HTTP para la solicitud y recuperación de datos de un servidor web, como páginas HTML, imagenes o archivos) o POST (es una solicitud HTTP que se usa para enviar datos desde un cliente a un servidor). Luego de que el servidor procese la solicitud, envía la respuesta al cliente, generalmente en formatos JSON o XML.

#endpoint: es una dirección URL específica dentro de una API que permite
#acceder a un recurso determinado. A través de este punto el cliente
#envía solicitudes al servidor para obtener o modificar información.

#¿Qué es el protocolo HTTP y cuál es su papel en la comunicación cliente-servidor?
#HTTP significa Hypertext Transfer Protocol, que básicamente es un conjunto de reglas estándarizadas de la capa de aplicación (o puentes que permiten que unos programas se comuniquen con la red) que estas rigen cómo se va a transferir la información entre el cliente o solicitante y un servidor web a través de internet.
#Esto funciona mediante una estructura de solicitud-respuesta, donde el navegador (cliente) solicite recursos y el servidor se los entregue.

#¿Qué es una petición HTTP y cómo funciona dentro del consumo de APIs?
#Dentro de las APIs este es un lenguaje ya estandarizado, que permiten que dos sistemas se comuniquen entre sí para intercambiar datos de manera estructurada.

#¿Cuáles son los principales métodos HTTP?
#Tenemos los siguientes:
#GET: Este es un método de solicitud que solamente es para solicitar o recuperar datos de un recurso específico en un servidor.
#POST: Este se usa para enviar datos al servidor con el fin de actualizar o crear un recurso.
#PUT: Este permite reemplazar completamente un recurso existente o crear en una URI específica cuando el cliente ya conoce su ubicación.
#DELETE: Este permite borrar un recurso del servidor cuando está identificado por una URI.

# desde la API https://jsonplaceholder.typicode.com/
API_URL = "https://jsonplaceholder.typicode.com/users"

#¿Qué son los códigos de estado HTTP?
#Los códigos de estado HTTP son respuestas numéricas que el servidor envía al cliente
#para indicar el resultado de una solicitud. Estos códigos informan si la petición
#fue exitosa, si ocurrió un error o si se requiere alguna acción adicional.

#Algunos ejemplos comunes son:
#200: Indica que la solicitud se procesó correctamente.
#404: Indica que el recurso solicitado no fue encontrado en el servidor.
#500: Indica un error interno del servidor.

def extraer_datos(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception(f"Error al conectar con la API: {response.status_code}")

#Ahora miremos esto de (.json)

##¿Qué es JSON?
#JSON (JavaScript Object Notation) es un formato ligero de intercambio de datos
#utilizado para almacenar y transmitir información entre aplicaciones.
#Es ampliamente usado en las APIs porque es fácil de leer para los humanos
#y sencillo de procesar por los lenguajes de programación.

#Los datos en JSON se organizan en pares clave-valor, similares a los diccionarios
#en Python.

#El método .json() de la librería Requests convierte automáticamente la respuesta
#de la API en formato JSON a una estructura de datos de Python, generalmente
#un diccionario o una lista de diccionarios, lo que permite manipular la
#información dentro del programa.

#¿Qué es una base de datos relacional y cuál es su utilidad en el almacenamiento de información?
#Es un tipo de base de datos que organiza la información en tablas relacionadas entre sí.
#Sirve para almacenar, organizar y consultar datos de forma eficiente.

# Creacion de la base de datos en SQLite "datos_api.db"
# se realiza la conexion a dicha base de datos
# y se crea el "cursor" para "execute" el CREATE TABLE y el INSERT OR REPLACE INTO usuarios

#¿Qué es SQLite y cuáles son sus principales características dentro del desarrollo de aplicaciones?
#SQLite es un sistema de gestión de bases de datos ligero que no necesita servidor.
# Es rápido, fácil de usar y guarda toda la base de datos en un solo archivo.

def guardar_en_db(datos):
    conn = sqlite3.connect('datos_api.db')
    cursor = conn.cursor()

    # Creación de la tabla "usuarios" dentro de la base de datos "datos_api.db"
    # esta contiene los campos id, nombre, usuario y email

    #¿Qué es una tabla dentro de una base de datos y qué función cumple una clave primaria (PRIMARY KEY)?
    # Una tabla es una estructura donde se almacenan datos en filas y columnas.
    # La clave primaria identifica de forma única cada registro dentro de la tabla.

    #¿Qué es SQL (Structured Query Language) y para qué se utiliza en bases de datos?
    # SQL es un lenguaje que se utiliza para crear, consultar, modificar y eliminar datos en bases de datos.

    #¿Qué función cumplen las instrucciones SQL CREATE TABLE, INSERT INTO y SELECT?
    # CREATE TABLE crea una nueva tabla en la base de datos.
    # INSERT INTO agrega nuevos registros a una tabla.
    # SELECT permite consultar o mostrar datos almacenados.

    #¿Cómo se insertan registros dentro de una tabla en una base de datos?
    #Se utiliza la instrucción INSERT INTO seguida del nombre de la tabla y los valores que se quieren agregar.

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS usuarios (
            id INTEGER PRIMARY KEY,
            nombre TEXT,
            usuario TEXT,
            email TEXT
        )
    ''')

    # Inserción de datos en la tabla "usuarios"
    for item in datos:
        cursor.execute('''
            INSERT OR REPLACE INTO usuarios (id, nombre, usuario, email)
            VALUES (?, ?, ?, ?)
        ''', (item['id'], item['name'], item['username'], item['email']))

    # Aceptación mediante "Commit" el ingreso de datos en la tabla "usuarios"
    conn.commit()

    # Cerrado de la conexion a la base de datos "datos_api.db"
    conn.close()

#¿Qué es la librería Pandas en Python y para qué se utiliza?
#Pandas es una librería de Python que se utiliza para analizar y manipular datos fácilmente.

#¿Qué es un DataFrame y por qué es útil para el análisis de datos?
# Un DataFrame es una estructura de datos en forma de tabla con filas y columnas.
# Es útil porque permite organizar y analizar grandes cantidades de datos de forma sencilla.

#¿Cómo se pueden exportar datos a archivos como CSV utilizando Pandas?
#Se utiliza el método to_csv() para guardar los datos de un DataFrame en un archivo CSV.

# 3. GENERAR EVIDENCIAS (Pandas y Auditoría)
def generar_evidencias(datos_api):
    # --- Archivo Excel/CSV con Pandas ---
    conn = sqlite3.connect('datos_api.db')
    df_db = pd.read_sql_query("SELECT * FROM usuarios", conn)
    df_db.to_csv('muestra_usuarios.csv', index=False)
    conn.close()

    # la variable "registros_api" obtiene la cantidad de registros recibidos desde la
    # API con url "https://jsonplaceholder.typicode.com/users"
    # la variable "registros_db" obtiene la cantidad de registros almacenados en la tabla "usuarios"
    registros_api = len(datos_api)
    registros_db = len(df_db)

    # Creación del archivo "auditoría.txt" con el metodo "open" de Python
    with open('auditoria.txt', 'w', encoding='utf-8') as f:
        f.write("RESUMEN DE AUDITORÍA\n")
        f.write("====================\n")
        f.write(f"Registros extraídos de la API https://jsonplaceholder.typicode.com/: {registros_api}\n")
        f.write(f"Registros guardados en la DB de SQLite3: {registros_db}\n")

        if registros_api == registros_db:
            f.write("\nESTADO: Operacion Exitosa.\n")
        else:
            f.write("\nESTADO: Operacion Fallida.\n")

# EJECUCIÓN DEL PROCESO

#¿Qué es el proceso ETL (Extract, Transform, Load) en sistemas de información?
# ETL es un proceso que consiste en extraer datos, transformarlos y cargarlos en una base de datos o sistema.

#¿Cómo se aplica el proceso ETL en un programa que obtiene datos desde una API y los almacena en una base de datos?
# Primero se extraen los datos desde la API, luego se transforman o limpian,
# y finalmente se cargan en la base de datos para su almacenamiento y análisis.

try:
    print("Iniciando proceso de petición y recepción de datos desde la API https://jsonplaceholder.typicode.com/users...")
    datos = extraer_datos(API_URL)

    print("Creando la base de datos datos_api.db y guardando datos en la tabla usuarios...")
    guardar_en_db(datos)

    print("Generando archivos de evidencia .CSV y auditoría.txt...")
    generar_evidencias(datos)

    print("¡Proceso completado con éxito!")
except Exception as e:
    print(f"Ocurrió un error: {e}")
```
